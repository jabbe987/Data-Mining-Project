```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
diabetes =  pd.read_csv("Diabetes.csv", delimiter=",")
diabetes.replace('no data', np.nan, inplace=True)

missing_values_count = diabetes.isnull().sum()
print(missing_values_count)

sns.heatmap(diabetes.isnull(), cbar=False, cmap='viridis')

plt.title('Heatmap of Missing Data in Diabetes Dataset')

# Show the plot
plt.show()
```

```python
# Load the diabetes dataset
diabetes = pd.read_csv("Diabetes.csv", delimiter=",")

# Replace 'no data' with NaN
diabetes.replace('no data', np.nan, inplace=True)

# Display the original number of rows
original_count = diabetes.shape[0]

diabetes = diabetes[~diabetes['Gender'].isin(['Males', 'Females'])]

removed_count = original_count - diabetes.shape[0]

print(f"Removed {removed_count} rows.")

# Drop rows that are missing a country name
diabetes.dropna(subset=['Country Name'], inplace=True)

columns_to_drop = ['Cancer Type', 'Area', 'Lower Confidence Limit', 
                   'Upper Confidence Limit', 'Comment', 'Unnamed: 11']  # Add any other unnecessary columns here

diabetes.drop(columns=columns_to_drop, axis=1, inplace=True)


# Reset the index of the DataFrame
diabetes.reset_index(drop=True, inplace=True)
print(diabetes.head)

sns.heatmap(diabetes.isnull(), cbar=False, cmap='viridis')

plt.title('Heatmap of Missing Data in Diabetes Dataset')

# Show the plot 
plt.show()
```
